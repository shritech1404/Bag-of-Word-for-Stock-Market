```python
data = pd.read_csv("/kaggle/input/stockmarket-sentiment-dataset/stock_data.csv")
data.head()
```

```python
X = data.Text
y = data.Sentiment
```

```python
X
```

```python
y
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
ps = PorterStemmer()
lammat = WordNetLemmatizer()
corpus = []
```

```python
for i in range(len(X)):
    review = re.sub('[^a-zA-Z]', " ", X[i])
    review = review.lower()
    review = review.split()
    ## review = [ps.stem(word)for word in review if not word in set(stopwords.words('english'))]
    review = [lammat.lemmatize(word)for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import numpy as np    
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_new = cv.fit_transform(corpus).toarray()
y_new = np.array(y)
```

```python
X_new
```

```python
y_new
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
